```python
import pyautogui
import pyttsx3
import requests
import speech_recognition as sr
import datetime
import os
import cv2
from requests import get
import wikipedia
import webbrowser
import pywhatkit
import smtplib
import sys
import pyjokes
import PyPDF2
import time
from bs4 import BeautifulSoup
import phonenumbers
from phonenumbers import geocoder
from phonenumbers import carrier
import pycountry
from phonenumbers.phonenumberutil import region_code_for_country_code
from phonenumbers.phonenumberutil import region_code_for_number
from opencage.geocoder import OpenCageGeocode
import folium


engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voices', voices[0].id)
engine.setProperty('rate', 187)


def speak(audio):
    engine.say(audio)
    print(audio)
    engine.runAndWait()

def command():
    x = sr.Recognizer()
    with sr.Microphone() as source:
        print("Please speak...")
        x.pause_threshold = 5
        audio = x.listen(source, timeout=5, phrase_time_limit=5)

    try:
        print("Processing...")
        query = x.recognize_google(audio, language='en-in')
        print(f"You said: {query}")

    except Exception as e:
        speak("I didn't quite get that.")
        return "none"
    return query

def greet():
    hour = int(datetime.datetime.now().hour)
    tt = time.strftime("%I:%M %p")

    if hour >= 0 and hour <= 12:
        speak(f"Good Morning! It is {tt}")
    elif hour>12 and hour <18:
        speak(f"Good Afternoon! It is {tt}")
    else:
        speak(f"Good Evening! It is {tt}")
    speak("How may i help you today!")

def news():
    main_url = 'http://newsapi.org/v2/top-headlines?sources=techcrunch&apiKey=1b75cb383b354f1ea63bca66256e5d3c'

    main_page = requests.get(main_url).json()
    articles = main_page["articles"]
    head = []
    day = ['first', 'second', 'third', 'fourth', 'fifth', 'sixth', 'seventh', 'eighth', 'ninth', 'tenth']
    for ar in articles:
        head.append(ar["title"])
    for i in range(len(day)):
        speak(f"today's {day[i]} news is: {head[i]}")

def pdf_reader():
    speak("Which file would you like to read?")
    file = input("Enter the file name: ")
    book = open(f"{file}", "rb")
    pdfReader = PyPDF2.PdfFileReader(book)
    pages = pdfReader.numPages
    speak(f"Total number of pages is {pages}")
    speak("Please enter the page number to be read out")
    pg = int(input("Enter page number: "))
    page = pdfReader.getPage(pg)
    text = page.extractText()
    speak(text)

if __name__ == "__main__":
    greet()
    while True:

        query = command().lower()

        if "open notepad" in query:
            path1 = "C:\\Windows\\System32\\notepad.exe"
            os.startfile(path1)

        elif "command prompt" in query:
            os.system("start cmd")

        elif "camera" in query:
            cap = cv2.VideoCapture(0)
            while True:
                ret, img = cap.read()
                cv2.imshow('webcam', img)
                k = cv2.waitKey(50)
                if k == 27:
                    break
                cap.release()
                cv2.destroyAllWindows()

        elif "ip address" in query:
            ip = get('https://api.ipify.org').text
            speak(f"Your IP address is {ip}")

        elif "wikipedia" in query:
            speak("searching wikipedia")
            query = query.replace("wikipedia", "")
            results = wikipedia.summary(query, sentences = 2)
            speak("I found this from wikipedia")
            speak(results)
            speak("Hope this helped!")

        elif "youtube" in query:
            speak("What would you like to play on youtube?")
            vid = command().lower()
            pywhatkit.playonyt(f"{vid}")

        elif "google" in query:
            speak("What would you like to search in google?")
            cm = command().lower()
            webbrowser.open(f"{cm}")

        elif "facebook" in query:
            webbrowser.open("www.facebook.com")

        elif "twitch" in query:
            webbrowser.open("www.twitch.com")

        elif "send message" in query:
            speak("who would you like to send the message to?")
            num = int(input("Enter phone number: "))
            speak("What message would you like to convey?")
            msg = command().title()
            pywhatkit.sendwhatmsg(f"+91{num}",f"{msg}", 21, 20)

        elif "close notepad" in query:
            speak("Closing notepad")
            os.system("taskkill /f /im notepad.exe")

        elif "joke" in query:
            joke = pyjokes.get_joke()
            speak(joke)

        elif "shut down" in query:
            ("Goodbye, hope to see you soon")
            os.system("shutdown /s /t 5")

        elif "restart" in query:
            speak("I will be back soon!")
            os.system("shutdown /r /t 5")

        elif "sleep" in query:
            speak("Good night!")
            os.system("rundll32.exe powrprof.dll, SetSuspendState 0,1,0")

        elif "switch window" in query:
            pyautogui.keyDown("alt")
            pyautogui.press("tab")
            pyautogui.keyUp("alt")

        elif "news" in query:
            speak("Here is some interesting news i found")
            news()

        elif "where am i" in query:
            try:
                ipAdd = requests.get("https://api.ipify.org").text
                print(ipAdd)
                url = "https://get.geojs.io/v1/ip/geo/" + ipAdd + '.json'
                geo_requests = requests.get(url)
                geo_data = geo_requests.json()
                city = geo_data["city"]
                country = geo_data["country"]
                speak(f"It appears that you are currently in {city} from {country}")
            except Exception as e:
                speak("Due to network issues, I am currently unable to do so!")
                pass

        elif "take screenshot" in query:
            speak("What would you like the screenshot to be named as?")
            name = command().lower()
            speak("Just a moment...")
            datetime.time.sleep(3)
            image = pyautogui.screenshot()
            img.save(f"{name}.png")
            speak("The screenshot has been taken.")

        elif "read pdf" in query:
            pdf_reader()

        elif "hide folder" in query:
            os.system("attrib +h /s /d")
            speak("All files are now hidden")

        elif "visible" in query:
            os.system("attrib -h /s /d")
            speak("All files have been made visible")

        elif "temperature" in query:
            speak("What is the temperature?")
            city = command().lower()
            search = f"temperature in {city}"
            url = f"https://www.google.com/search?q={search}"
            m = requests.get(url)
            data = BeautifulSoup(m.text, "html.parser")
            temp = data.find("div", class_="BNeawe").text
            speak(f"current {search} is {temp}")

        elif "track" in query:
            if os.path.exists("mylocation.html"):
                os.remove("mylocation.html")
            try:
                number = input("Enter Phone Number: ")
                pn = phonenumbers.parse(number)

                country = pycountry.countries.get(alpha_2=region_code_for_number(pn))
                location = country.name
                print(location)

                print(carrier.name_for_number(phonenumbers.parse(number), "en"))

                key = input("Enter Your API KEY: ")                   #83f883a3c01d4f8bb39ecc40f90a56af
                geocoder = OpenCageGeocode(key)
                query = str(location)
                results = geocoder.geocode(query)
                lat = results[0]['geometry']['lat']
                lng = results[0]['geometry']['lng']
                print(lat, lng)

                myMap = folium.Map(location=[lat, lng], zoom_start=9)
                folium.Marker([lat, lng], popup=location).add_to(myMap)
                myMap.save("mylocation.html")
                os.system("mylocation.html")
            except:
                print("Missing or invalid number\nPlease enter your phone number with country code")

        elif "hi" in query:
            speak("Hello")

        elif "how are you" in query:
            speak("I am great thank you. How about you?")

        elif "i am bored" in query:
            speak("Would you like to watch some movies?")
            a1 = command().lower()
            if "yes" in a1:
                speak("Okay! lets watch a movie.")
                webbrowser.open("https://ww5.soap2dayhd.co/home/")
            elif "no" in a1:
                speak("Perhaps some game?")
                a2 = command().lower()
                if "yes" in a2:
                    speak("Great!")
                    webbrowser.open("https://www.friv.com")
                elif "no" in a2:
                    speak("What else would you like to do?")
                    a3 = command().lower()
                    speak("Sure thing!")
                    webbrowser.open(f"{a3}")

        elif "no thanks" in query:
            speak("Thank you. Goodbye")
            sys.exit()

        speak("Would you like anything else?")
```
