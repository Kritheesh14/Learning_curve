```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import os
```

```python
def pltgraph(x, y, xlbl, ylbl, head, gridt, pdfc, pdfn):
    try:
        if not x or not y:
            raise ValueError("x and y lists cant be empty")
            
        plt.plot(x, y)
        plt.xlabel(xlbl)
        plt.ylabel(ylbl)
        plt.title(head)
        plt.grid(gridt)
        if pdfc:
            try:
                dpath=os.path.join(os.path.expanduser("~"), "Downloads")
                fullpath=os.path.join(dpath, pdfn)
                with PdfPages(fullpath) as pdf:
                    fig=plt.gcf()
                    pdf.savefig(fig, bbox_inches='tight')
                    plt.close(fig)
                print(f"Plot saved to {fullpath} (Downloads)")
            except Exception as e:
                print(f"Error saving PDF: {e}")
        plt.show()

    except ValueError as e:
        print(f"Error: {e}")
```
